```python
import numpy as np
import pandas as pd
```

```python
#**Consider the following Python dictionary data and Python list labels:**

data = {'birds': ['Cranes', 'Cranes', 'plovers', 'spoonbills', 'spoonbills', 'Cranes', 'plovers', 'Cranes', 'spoonbills', 'spoonbills'],
        'age': [3.5, 4, 1.5, np.nan, 6, 3, 5.5, np.nan, 8, 4],
        'visits': [2, 4, 3, 4, 3, 4, 2, 2, 3, 2],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
#  (1) Creating a DataFrame Birds from the Dictionary and indexing labels.

df = pd.DataFrame(data, columns=["birds", "age", "visits", "priority"])
print(df)
```

```python
# (2) Displaying the summary of the basic information about birds DataFrame and its Data.

df.describe()
```

```python
# (3) The first two rows of the Birds DataFrame.

df.head(2)
```

```python
# (5) Selecting [2,3,7] rowsand in columns['birds','age','visits']

df.loc[[2,3,7],['birds', 'age', 'visits']]
```

```python
#  (7) selecting the rows with columns ['birds', 'visits'] where the age is missing i.e NaN**

df = pd.DataFrame(data, columns = ['birds','age','visits'])
df[df['age'].isnull()]
```

```python
# (8). Select the rows where the birds is a Cranes and the age is less than 4**

options = ['Cranes']
df[(df['age']<4) & df['birds'].isin(options)]
```

```python
# (9). Select the rows the age is between 2 and 4(inclusive)**

df[(df['age'] >2 ) & (df['age'] < 5)]
```

```python
# (10) The total number of visits in the birds 'Cranes'

df['visits'].sum(axis=0)
```

```python
# (11). Calculate the mean age for each different birds in dataframe.**

df.groupby(['birds'])['age'].mean()
```

```python
# (13). Find the number of each type of birds in dataframe (Counts)**

df['birds'].value_counts()
```

```python
'''(14) Sort dataframe (birds) first by the values in the 'age' in decending order, then by the value in the 'visits' column 
in ascending order.'''

df.sort_values(by=['age', 'visits'], ascending = [False,True])
```
